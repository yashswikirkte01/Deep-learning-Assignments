```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN,Dense
```

```python
x = np.linspace(0,100,1000) #creates a numpy array of 1000 evenly spaced values between 0 and 100.
y = np.sin(x)
# Creates a sine wave.This acts as our time series data
```

```python
timesteps = 10
X,Y = [],[]
for i in range(len(y)-timesteps):
    X.append(y[i:i+timesteps])
    Y.append(y[i+timesteps])

    # X stores the past 10 values input
    # Y stores the next value
```

```python
X
```

```python
Y
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
X = X.reshape((X.shape[0],X.shape[1],1))
```

```python
model = Sequential([
    SimpleRNN(50,activation='tanh',input_shape=(timesteps,1)),
    Dense(1)
])
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X, Y, epochs=10, batch_size=32)
```

```python
pred = model.predict(X[:5])
print("Actual:", Y[:5])
print("Predicted:", pred.flatten())
```
